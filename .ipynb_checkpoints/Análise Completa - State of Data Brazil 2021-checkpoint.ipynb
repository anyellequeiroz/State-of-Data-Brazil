A State of Data Brazil foi realizada entre 18 de outubro de 2021 e 6 de dezembro de 2021 atravs de um questionrio online e reuniu indicadores relacionados a **perfil demogrfico, formao, atuao no setor, remunerao, rotatividade e fatores de satisfao no ambiente de trabalho,** incluindo o impacto do trabalho remoto nas preferncias profissionais de 2.645 respondentes de todo o Brasil. A amostra reflete a viso de variados papis de atuao em empresas, como os de analista de dados, cientista de dados e engenheiro de dados, bem como diferentes perfis de experincia profissional, incluindo analistas jnior, pleno, snior e gestores.

H uma grande quantidade de colunas e elas so divididas em 10 categorias, de P0 a P9. Cada categoria  relacionada com um tipo de anlise diferente.

- **P0: Id de cada usurio que respondeu ao questionrio.**
- **P1: Perguntas pessoais e demogrficas: Idade, Estado, Genero, ... df_1**
- **P2: Perguntas relacionadas ao emprego atual: Salario, Cargo, Nvel, ... df_2**
- **P3: Perguntas mais especficas sobre o emprego atual: Nvel da equipe, ... df_3**
- **P4: Perguntas sobre como  a analise de dados na empresa, que tipo de dado  utilizado. df_4**
- **P5: Perguntas sobre novas oportunidades (Apenas 3 perguntas) df_5**
- **P6: Perguntas sobre rotina de trabalho relacionada a Engenharia de Dados e ferramentas utilizadas df_6**
- **P7: Perguntas sobre rotina de trabalho relacionada a Anlise de Dados e ferramentas utilizadas df_7**
- **P8: Perguntas sobre rotina de trabalho relacionada a Ciencia de Dados e ferramentas utilizadas df_8**
- **P9: Perguntas sobre interao com Data Hackers df_9**

*Vamos entender cada uma das categorias e obter informaes relevantes sobre o estudo.*

# P1: Questes Pessoais e Demogrficas

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

data = pd.read_csv('State of Data.csv')

df_1 = data.loc[:,"('P1_a ', 'Idade')":"('P1_i ', 'rea de Formao')"]
df_1.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_2 = data.loc[:,"('P2_a ', 'Qual sua situao atual de trabalho?')":"('P2_s ', 'Caso sua empresa decida pelo modelo 100% presencial qual ser sua atitude?')"]
df_2.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_3 = data.loc[:,"('P3_a ', 'Qual o nmero aproximado de pessoas que atuam com dados na sua empresa hoje?')":"('P3_d_n ', 'Dividir o tempo entre entregas tcnicas e gesto.')"]
df_3.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_4 = data.loc[:,"('P4_a_a ', 'Mesmo que esse no seja seu cargo formal, voc considera que sua atuao no dia a dia, reflete alguma das opes listadas abaixo?')":"('P4_h_x ', 'No utilizo nenhuma ferramenta de BI no trabalho')"]
df_4.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_5 = data.loc[:,"('P5_b ', 'Qual oportunidade voc est buscando?')":"('P5_d ', 'Como tem sido a busca por um emprego na rea de dados?')"]
df_5.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_6 = data.loc[:,"('P6_a ', 'Quais das opes abaixo fazem parte da sua rotina no trabalho atual como engenheiro de dados?')":"('P6_h_h ', 'Cuidando da qualidade dos dados, metadados e dicionrio de dados.')"]
df_6.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_7 = data.loc[:,"('P7_a ', 'Quais das opes abaixo fazem parte da sua rotina no trabalho atual com anlise de dados?')":"('P7_d_h ', 'Utilizando ferramentas avanadas de estatstica como SAS, SPSS, Stata etc, para realizar anlises.')"]
df_7.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_8 = data.loc[:,"('P8_a ', 'Quais das opes abaixo fazem parte da sua rotina no trabalho atual com cincia de dados?')":"('P8_d_i ', 'Criando e mantendo a infra que meus modelos e solues rodam (clusters, servidores, API, containers, etc.)')"]
df_8.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])

df_9 = data.loc[:,"('P9_a ', 'Quais das iniciativas do Data Hackers que voc j acessou/acompanhou?')":"('P9_a_g ', 'Ainda no conhecia o Data Hackers')"]
df_9.insert(loc=0, column="('P0', 'id')", value=data["('P0', 'id')"])
```

```python
ax = sns.histplot(df_1["('P1_a ', 'Idade')"]);
ax.set_title('Distribuio de idades dentre as pessoas que responderam a pesquisa.')
ax.figure.set_size_inches(12, 8)
ax.set_ylabel('Quantidade de pessoas')
ax.set_xlabel('Idades');
```

```python
ax = sns.countplot(df_1["('P1_b ', 'Genero')"]);
ax.set_title('Distribuio de gneros dentre as pessoas que responderam a pesquisa.')
ax.figure.set_size_inches(12, 8)
ax.set_ylabel('Quantidade de pessoas')
ax.set_xlabel('Gneros');
```

```python
import geobr 

count_states = df_1.groupby(by=["('P1_e_a ', 'uf onde mora')"]).count().reset_index()
states = geobr.read_state(year=2019, simplified=True)
states = states.merge(count_states, how="left", left_on="abbrev_state", right_on="('P1_e_a ', 'uf onde mora')")
states["('P0', 'id')"] = states["('P0', 'id')"].fillna(0)

plt.rcParams.update({"font.size": 5})

fig, ax = plt.subplots(figsize=(8, 8), dpi=200)

states.plot(
    column="('P0', 'id')",
     cmap='OrRd',
     legend=True,
     legend_kwds={
         "shrink": 0.6
     },
     ax=ax
)
ax.set_title("Onde esto as pessoas que trabalham na rea?")
ax.axis("off");
```
